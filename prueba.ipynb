```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('./data/forecast.csv')
```

```python
# Crear la pivot table
pivot_df = df.pivot_table(values='sales', index='item', columns='date', fill_value=0)
```

```python
pivot_df.head()
```

```python
# Crear la pivot table
pivot_df = df.pivot_table(values='sales', index='item', columns='date', aggfunc='sum', fill_value=0)
```

```python
# Agregar la columna 'Total' que es la suma de todas las ventas por fila
pivot_df['Total'] = pivot_df.sum(axis=1)
```

```python
pivot_df.head()
```

```python
# Obteniendo los items únicos
items = df['item'].unique()

# Generando datos aleatorios para el stock
np.random.seed(0)  # Para reproducibilidad de los resultados
stocks = np.random.randint(10, 100, size=len(items))

# Creando el DataFrame
df_stock = pd.DataFrame({
    "item": items,
    "stock": stocks
})
```

```python
df_stock.head()
```

```python
pivot_df = pivot_df.merge(df_stock, on = 'item')
```

```python
pivot_df.head()
```

```python
pivot_df['Pedido'] = pivot_df['stock'] - pivot_df['Total']
```
